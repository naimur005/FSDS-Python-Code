### 01 - Temperature Check

```python
temperature = float(input("Please, enter the current temperature"))

if temperature < 10:
    print ("It's too cold!")
elif 10 <= temperature >= 25:
    print ("It's too cold outside")
else:
    print ("It's hot outside!")
```

### 03 - Student Grade Calculator

```python
marks = float(input("Please enter your number"))

if 90 <= marks <=100:
    print ("A+")
elif 80 <= marks <= 89:
    print ("A")
elif 70 <= marks <= 79:
    print ("B")
elif 60 <= marks <= 69:
    print ("C")
else:
    print ("Fail")
```

### 04 - Bus Ticket Discoutnt

```python
age = int(input("Please, enter your age"))

if age < 5:
    print("Free")
elif 5 <= age <= 18:
    print(" 50% Discount")
elif 19 <= age <= 60:
    print("30% Discount")
else:
    print("Full Price")
```

### 05 - Restaurant Bill with Discount

```python
bill = float(input("Please, enter the total bill amount"))

if bill > 1000:
    discount = bill * 0.10
    final_bill = bill - discount
    print(f"Final Bill after 10% discount: {final_bill}")
else:
    print(f"Final Bill: {bill}")
```

### 07 - Traffic Light System

```python
color = input("Please enter a color (red, yellow, or green)")

if color == "red":
    print("Stop")
elif color == "yellow":
    print("Ready to go")
elif color == "green":
    print("Go")
else:
    print("Invalid color")
```

### 08 - Leap Year Checker

```python
year = int (input ("Please enter a year"))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

### 09 - Job Eligibility Center

```python
age = int(input("Please, enter your age"))
qualification = input("Please, enter your qualification").lower()

if age >= 18 and qualification == "graduate":
    print("Eligible for Job")
else:
    print("Not Eligible")
```

### 10 - Electricity Bill Calculator

```python
units = int(input("Please, enter the number of units consumed"))

if units <= 100:
    bill = units * 5
elif 101 <= units <= 200:
    bill = units * 7
else:
    bill = units * 10

print(f"Total payable amount: {bill} BDT")
```
